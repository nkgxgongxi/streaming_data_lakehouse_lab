```python
from snowflake.snowpark.session import Session
import snowflake.snowpark.functions as F
import configparser
import os
from datetime import date
```

```python
def snowpark_session_create(config_dict):
    connection_params = {
        "account": config_dict['account'],
        "user": config_dict['user'],
        "password": config_dict['password'],
        "role": config_dict['role'],
        "warehouse": config_dict['warehouse']
    }

    session = Session.builder.configs(connection_params).create()

    return session
```

```python
script_dir = os.path.dirname(os.getcwd())
config_path = os.path.join(script_dir, 'snowpark_test', 'config.ini')
print(config_path)
config = configparser.ConfigParser()
config.read(config_path)
config_dict = (config['snowflake'])
demo_session = snowpark_session_create(config_dict=config_dict)
```

```python
df = demo_session.sql('select * from raw.financial_info.news_clean limit 5')
```

```python
# show - print version
# collect - into objects into rows where you can iterate (it's an object in your memory)
df.show()
for row in df.collect():
    print(row.URL)
```

```python
demo_session.use_database("raw")
demo_session.use_schema("financial_info")
```

```python
df = demo_session.table("news_clean")

df = df.select(
    "TITLE", 
    "CONTENT",
    "LAST_UPDATED_AT"
).filter((F.col("LAST_UPDATED_AT").between(date(2025,3,15), date(2025,3,16))))

df.show()
```
